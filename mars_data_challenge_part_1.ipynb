```python
from splinter.browser import Browser
import pandas as pd
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://redplanetscience.com/"
```

```python
browser.visit(url)
```

```python
html = browser.html
```

```python
soup1 = soup(html, 'html.parser')
soup1
```

```python
soup1.find('title').text
```

```python
tag_box = soup1.find('div', class_='content_title').text
tag_box
```

```python
python1 = {}
python1['title'] = tag_box
python1
```

```python
tag_article = soup1.find('div', class_='article_teaser_body').text
tag_article
```

```python
python2 ={}
python2['preview'] = tag_article
python2
```
